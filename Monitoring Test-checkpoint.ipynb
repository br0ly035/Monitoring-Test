```python
df3 = df3.fillna(0)
df3['minute'] = df3['minute'].str[:5]
df3[df3.columns[1:]] = df3[df3.columns[1:]].astype(int)
df3['apr_mean_15m'] = df3['approved'].rolling(window = 15).mean().round(2)
```

```python
pd.set_option('max_rows',10)
```

```python
@interact(start_time = widgets.IntSlider(min = 0, max = 18, step = 1, value = 0),
          upper_pct = widgets.IntSlider(min = 0, max = 100, step = 5, value = 0),
          lower_pct = widgets.IntSlider(min = 0, max = 100, step = 5, value = 0))
def approved_alert(upper_pct, lower_pct, start_time):
    temp_df = df3[['minute','approved','apr_mean_15m']].copy()
    temp_df['apr_mean_15m_lower'] = temp_df['apr_mean_15m'] * (1 - (lower_pct/100))
    temp_df['apr_mean_15m_upper'] = temp_df['apr_mean_15m'] * (1 + (upper_pct/100))
    alert_min_apr = temp_df[(temp_df.approved < temp_df.apr_mean_15m_lower) & (temp_df.minute.str[:2].astype(int) >= start_time)]
    alert_max_apr = temp_df[(temp_df.approved > temp_df.apr_mean_15m_upper) & (temp_df.minute.str[:2].astype(int) >= start_time)]
    temp_df.plot(x = 'minute', figsize = (10,10))
    plt.show()
    print(f"LOWER BOUND ALLERT: {len(alert_min_apr)}")
    display(alert_min_apr)
    print(f"UPPER BOUND ALLERT: {len(alert_max_apr)}")
    display(alert_max_apr)
```
